```python
# Column indexes were changed by Daniel K and Alison ..Date Jun 11 2020
# Education and Age related columns added on June 11..validation pending
import pandas as pd
import csv

def mort2018Parser( line ):
	ret_dict = dict(
		Resident_status = line[19], # Index correct
		Place_of_death_and_decedents_status = line[82], # Index correct
		Day_of_week_of_death = line[84],
        Manner_of_Death = line[106],
		Data_year = line[101:105],    # Corrected..SM
		Month_of_Death = line[64:65],
		Sex = line[68], # Index correct
		Race = line[444:449],
		Age = line[69:81], # Can be dropped in the final stage after QA
        Age_Detail_Age = line[69:72],#SM..June 11..child of Age column .Confirm with team about naming convention
        Age_Age_Substitution_Flag = line[73], #SM..June 11..child of Age column .Confirm with team about naming convention
        Age_Age_Recode_52 = line[74:75],#SM..June 11..child of Age column .Confirm with team about naming convention 
        Age_Age_Recode_27 = line[76:77], #SM..June 11..child of Age column .Confirm with team about naming convention
        Age_Age_Recode_12 = line[78:79],#SM..June 11..child of Age column .Confirm with team about naming convention
        Age_Infant_Age_Recode_2 = line[80:81], #SM..June 11..child of Age column .Confirm with team about naming convention
        Age_Infant_Age_Recode_Code = line[82],#SM..June 11..child of Age column .Confirm with team about naming convention 
		Marital_status = line[83],
		Hispanic_origin = line[483:487],
		Education = line[60:63], # Can be dropped in the final stage after QA
        Education_1989_revision = line[60:61],#SM..June 11..child of Education column .Confirm with team about naming convention 
        Education_2003_revision = line[62],#SM..June 11..child of Education column .Confirm with team about naming convention 
        Education_Reporting_Flag = line[63],#SM..June 11..child of Education column .Confirm with team about naming convention 
		ICD_10_code = line[145:148],
		cause_recode_358 = line[149:151],
		cause_recode_113 = line[153:155],
		cause_recode_130 = line[156:158],
		cause_recode_39 = line[159:160],
		Entity_Axis_conditions_Number_of = line[162:163],
		Entity_Axis_conditions_conditions = line[164:303],
		Rec_axis_conditions_numberof = line[340:341],
		Rec_axis_conditions_conditions = line[343:442],
		Place_of_injury = line[144],
		Injury_at_work = line[105],
		Method_of_disposition = line[107],
		Autopsy = line[108],
		Activity_code = line[143]
	)
	return pd.Series( ret_dict )

def convert_to_csv( input_fp, output_fp ):
    with open( output_fp, 'w' ) as fout:
        writer = csv.writer( fout )
        with open( input_fp, 'r' ) as fin:
            line = fin.readline()
            row = mort2018Parser( line )
            writer.writerow( list(row.index) )
            while line:
                row = mort2018Parser( line )
                writer.writerow( list(row) )
                line = fin.readline()
    return output_fp
```
